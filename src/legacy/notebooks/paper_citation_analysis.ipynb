````python
import arxiv

def download_arxiv_paper(arxiv_id, output_filename=None):
    """Downloads an arXiv paper by ID and optionally saves it with a custom filename.

    Args:
        arxiv_id (str): The ID of the arXiv paper (e.g., "2203.01234v1").
        output_filename (str, optional): The desired filename for the downloaded paper.
            If not provided, a default filename will be used based on the paper's ID.

    Returns:
        arxiv.Result: The downloaded arXiv paper object.
    """

    client = arxiv.Client()
    search = arxiv.Search(id_list=[arxiv_id])
    paper = next(client.results(search))

    if output_filename:
        filename = output_filename
    else:
        filename = f"{paper.id}.pdf"  # Use default filename with ID

    paper.download_pdf(filename=filename)
    return paper

# Example usage:
downloaded_paper = download_arxiv_paper("1706.03762", output_filename="../data/papers/attention-is-all-you-need.pdf")
````

````python
import re

def expand_citations(text):
    # Regular expression to find citation patterns including ranges and lists
    citation_pattern = re.compile(r'\[(\d+(-\d+)?(,\s*\d+(-\d+)?)*)\]')

    expanded_citations = []

    for match in citation_pattern.finditer(text):
        # Extract the citation text without brackets
        citation_text = match.group(1)

        # Split the citation text by comma
        parts = citation_text.split(',')

        for part in parts:
            part = part.strip()  # Remove leading and trailing whitespaces

            # Check for a range (e.g., 1-3)
            if '-' in part:
                start, end = map(int, part.split('-'))
                # Expand the range
                expanded_citations.extend([i for i in range(start, end + 1)])
            else:
                # No range, just a single number
                expanded_citations.append(int(part))

    return expanded_citations

def extract_marked_sentences(elements, mark_pattern='\\[\\d+\\]'):
    """
    Extracts sentences from a list of text strings that contain a specific mark pattern.

    Args:
    text_list (list of str): The list of text strings to search through.
    mark_pattern (str): The regex pattern that defines the mark to search for. Default is '\\[\\d+\\]', which matches [number].

    Returns:
    list of str: A list of sentences that contain the specified mark.
    """
    import re

    # Compile the regular expression pattern for efficiency
    pattern = re.compile(mark_pattern)

    # List to hold all sentences with the specified mark
    marked_sentences = []

    # Iterate through each text string in the list
    for el in elements:
        
        # Split the text into sentences
        sentences = re.split(r'(?<=[.!?]) +', el.text)

        # Check each sentence for the specified mark
        for sentence in sentences:
            if pattern.search(sentence):
                marked_sentences.append(el)

    return marked_sentences
````

````python
from collections import defaultdict

# Function to split and group data by citation number
def split_and_group(data):
    grouped_data = defaultdict(lambda: {'Categories': set(), 'Explanation': ''})
    
    for item in data:
        # Extract citation numbers from the string, assuming they are integers
        citations = [int(c) for c in item['Citation'].strip('[]').split(', ')]
        
        for citation in citations:
            grouped_data[citation]['Categories'].add(item['Category'])
            grouped_data[citation]['Explanation'] += item['Explanation'] + ' '
    
    # Convert the set of categories to a list and trim trailing spaces from explanations
    for citation, details in grouped_data.items():
        details['Categories'] = list(details['Categories'])
        details['Explanation'] = details['Explanation'].strip()
    
    return grouped_data

grouped_data = split_and_group(data)
````
